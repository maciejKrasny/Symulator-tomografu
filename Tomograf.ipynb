```python
import sys
import os
import matplotlib.pyplot as plt
import cv2
import numpy as np
sys.path.append(os.getcwd())
import copy
import skimage
from skimage.transform import iradon, rescale

from tomograph import Scanner
from Plotter import Plotter
```

```python
image = skimage.io.imread('images/Kwadraty2.jpg', as_gray=True)
normalized_image = np.interp(image, (image.min(), image.max()), (0, 255))
```

```python
scanner = Scanner(steps=360, detectors_count=200, spread=270)
sinogram = scanner.create_sinogram(normalized_image)
```

```python
ramp_filter = scanner.filter(sinogram)
Plotter.plotSinogram(ramp_filter)
```

```python
deconstructed = scanner.deconstruct_sinogram(ramp_filter)
```
